```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fund_sentiment_data.csv')
```

```python
correlation = df[['Fund_Price', 'Negative_Sentiment_Count', 'Positive_Sentiment_Count']].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Fund Price and News Sentiments")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Fund_Price', data=df, label='Fund Price', color='blue')
sns.lineplot(x='Date', y='Negative_Sentiment_Count', data=df, label='Negative News Volume', color='red')
plt.xticks(rotation=45)
plt.title('Fund Price vs. Negative News Volume Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load data
df = pd.read_csv("fund_sentiment_data.csv")

# Example: Daily Pareto between Negative Sentiment and Fund Price
fig = go.Figure()

# Bar for negative sentiment
fig.add_trace(go.Bar(
    x=df['Date'],
    y=df['Negative_Sentiment_Count'],
    name='Negative Sentiment Count',
    marker=dict(color='indianred'),
    yaxis='y1'
))

# Line for fund price
fig.add_trace(go.Scatter(
    x=df['Date'],
    y=df['Fund_Price'],
    name='Fund Price',
    mode='lines+markers',
    yaxis='y2',
    line=dict(color='royalblue')
))

# Layout settings
fig.update_layout(
    title='Pareto Analysis: Negative News vs Fund Price',
    xaxis=dict(title='Date'),
    yaxis=dict(title='Negative Sentiment Count', side='left'),
    yaxis2=dict(title='Fund Price', overlaying='y', side='right'),
    legend=dict(x=0.01, y=0.99),
    template='plotly_dark',
    height=500,
    width=900
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load CSV
df = pd.read_csv("fund_sentiment_data.csv")

# Create graph
fig = go.Figure()

# Bar chart for Negative Sentiment
fig.add_trace(go.Bar(
    x=df['Date'],
    y=df['Negative_Sentiment_Count'],
    name='Negative Sentiment Count',
    marker=dict(
        color='rgba(255, 99, 71, 0.7)',
        line=dict(color='rgba(255, 99, 71, 1.0)', width=1.5)
    ),
    yaxis='y1'
))

# Line chart for Fund Price
fig.add_trace(go.Scatter(
    x=df['Date'],
    y=df['Fund_Price'],
    name='Fund Price',
    mode='lines+markers',
    line=dict(color='rgba(30, 144, 255, 0.9)', width=3),
    marker=dict(size=8, symbol='circle', color='white',
                line=dict(width=2, color='rgba(30, 144, 255, 1.0)')),
    yaxis='y2'
))

# Elegant Layout
fig.update_layout(
    title=dict(
        text='📊 Elegant Pareto Analysis: Fund Price vs Negative Sentiment Volume',
        x=0.5,
        font=dict(size=22, family='Verdana', color='white')
    ),
    xaxis=dict(
        title='Date',
        tickangle=45,
        tickfont=dict(size=12, color='white'),
        titlefont=dict(size=16, color='white'),
        showgrid=False
    ),
    yaxis=dict(
        title='Negative Sentiment Count',
        titlefont=dict(color='rgba(255, 99, 71, 1.0)', size=16),
        tickfont=dict(color='rgba(255, 99, 71, 1.0)'),
        showgrid=False
    ),
    yaxis2=dict(
        title='Fund Price',
        titlefont=dict(color='rgba(30, 144, 255, 1.0)', size=16),
        tickfont=dict(color='rgba(30, 144, 255, 1.0)'),
        overlaying='y',
        side='right',
        showgrid=False
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1,
        font=dict(color='white')
    ),
    plot_bgcolor='rgba(17,17,17,1)',
    paper_bgcolor='rgba(17,17,17,1)',
    margin=dict(t=80, b=100),
    height=600,
    width=1000
)

fig.show()
```
